Το προηγούμενο σύστημα συστάσεων προτείνει στον χρήστη ταινίες αποκλειστικά με βάση την ομοιότητά τους. Σε αυτό το σημείο θα γίνει μια αναβάθμιση του συστήματος έτσι ώστε να παράγονται καλύτερες  συστάσεις οι όποιες θα λαμβάνουν υπόψιν και τις προτιμήσεις του χρήστη, οι όποιες θα εξάγονται από τις αξιολογήσεις που έχει κάνει μέχρι στιγμής. 

Η εκπαίδευση του recommender θα γίνεται ως εξής:

Για κάθε ταινία θα υπάρχει ένα score το όποιο αρχικά θε είναι ίσο με 0 και θα διαμορφώνεται από τις αξιολογήσεις κάθε user. Έτσι για έναν χρήστη με βάση τις αξιολογήσεις που υπάρχουν στο αρχείο train_ratings θα πρέπει:

1.	Για κάθε ταινία που έχει βαθμολογήσει να βρίσκονται οι κοινές ταινίες ανά κλίμακα και στο μέχρι τώρα σκορ κάθε παρόμοιας ταινίας θα προστίθεται ένα βάρος το όποιο θα μπορούσε να είναι το ποσοστό ομοιότητας της ταινίας (δηλαδή ένα βάρος για κάθε κλίμακα) επί τον βαθμό που έχει βάλει ο χρήστης για την αρχική ταινία. Στη συνέχεια, ανάλογα με το σκορ που έχει σχηματιστεί για κάθε ταινία, θα επιλέγεται αν αυτή θα μπορούσε να είναι προτεινόμενη για τον χρήστη.

Η λογική πίσω από την παραπάνω διαδικασία είναι ότι παρόμοιες ταινίες θα έχουν ανάλογο βαθμό. Για παράδειγμα αν ένας χρήστης έχει αξιολογήσει αρκετές  ταινίες οι όποιες είναι sci-fiction με 5/5 τότε μια ταίνια sci-fiction την όποια δεν έχει δεί λογικά θα του αρέσει και θα έπρεπε να την προτείνουμε.

Μετά την εκπαίδευση του συστήματος σας καλείστε να δοκιμάσετε τον recommender που κατασκευάσατε στην πράξη. Για τον σκοπό αυτό θα φορτώσετε το αρχείο train_ratings.csv όπου περιέχονται οι αξιολογήσεις του ίδιου χρήστη για άλλες ταινίες. Το σύστημά σας πρέπει να προβλέπει αν μια ταινία θα πρέπει να προταθεί στον χρήστη. Μια ταίνια θα έπρεπε να έχει προταθεί στον  χρήστη, αν έχει βαθμό μεγαλύτερο του 3. Συνεπώς για την επίβλεψη του συστήματός σας θα πρέπει για κάθε μια από τις ταινίες του αρχείου test_ratings.csv να επιστρέφεται 1 ή 0 αν η ταινία θα έπαιρνε βαθμό μεγαλύτερο του 3 ή όχι, δηλαδή αν θα έπρεπε να την είχαμε προτείνει ή όχι.

Στην συνέχεια, σε συνδιασμό με τις πραγματικές απαντήσεις του χρήστη θα αξιολογήσετε το σύστημά σας χρησιμοποιόντας τις μετρικές: precision, recall, f1 οι οποίες είναι οι πλέον γνωστές μετρικές και ευρέως χρησιμοποιούμενες τεχνικές για την επίβλεψη-μέτρηση απόδοσης ανάλογων συστημάτων.

1. Precision: Δείχνει πόσο ακριβές είναι το σύστημα. Υπολογίζει πόσα από τα στιγμιότυπα τα όποια προβλέψαμε ότι ανήκουν σε μια κλάση όντως ανήκουν σε αυτή. Η μετρική αυτή μας δίνει μια εικόνα σχετικά με τον αριθμό των ταινίων που προβλέψαμε ως προτεινόμενες ενώ δεν θα έπρεπε.

2. Recall: Υπολογίζει πόσα από τα στιγμιότυπα που ανήκουν σε μια κλάση (π.χ. προτεινόμενες ταινίες) προβλέφθηκαν σωστά.

3. F1: Είναι ένας μέσος μεταξύ των παραπάνω δυο μετρικών, έτσι ώστε να διατηρείται μια ισορροπία μεταξύ τους. Υπολογίζεται από την παρακάτω σχέση:

$$F_1=2\frac{Precision\times{Recall}}{Precision+Recall}$$

Οι παραπάνω συναρτήσεις παρέχονται από την βιβλιοθήκη scikit-learn.

Τέλος για την καλύτερη επίβλεψη του συστήματος σας μπορείτε να εκπαιδεύσετε  τον recommender σας με ένα υποσύνολο ταινίων (π.χ. 10, 20, 50, ...) για να μελετήσετε κατά πόσο σας βοηθούν οι επιπλέον αξιολογήσεις κάθε φορά (δηλαδή κατά πόσο βελτιώνονται οι παραπάνω μετρικές στο test set). Έτσι π.χ. μπορείτε να εντοπίσετε περιπτώσεις όπως για παράδειγμα ότι με έναν recommender μπορεί να μην επιτυγχάνετε πολύ υψηλό σκορ όσο με άλλους, αλλά το βέλτιστο σκόρ σας επιτυγχάνεται πολύ γρηγόρα π.χ. με μόνο 10 ταινίες αντί 100.

````python
import pandas as pd
from pyswip import Prolog
````

````python
# Η βιβλιοθήκη pandas είναι χρήσιμη για την εργασία με τέτοια δεδομένα
import pandas as pd 
# Διάβασμα του αρχείου 'movie_metadata.csv' 
data = pd.read_csv(path + "movies_metadata.csv") 
#data = pd.read_csv("movies_metadata.csv") 

#Στο csv υπαρχούν κελία με nan τιμές
#Στις θέσεις αυτές βάζουμε 'UNK' πράγμα που  κάνουμε με την παρακάτω συνάρτηση
data.fillna("UNK", inplace=True)
# Preview the first 5 lines of the loaded data 
data.head()
````

````python
def clean_text(text):
  text = text.replace(u'\xa0', u'')
  text = text.replace(u"'", u'')
  text = text.replace(u'"', u'')
  return text
````

````python
#create World
#Ορίζουμε τον κόσμο μας
prolog = Prolog()

#1α

#Για κάθε row του πίνακα φτιάχνουμε τα κατηγορήματα που θέλουμε να αποθηκέυσουμε
#αρχικά σε μια λίστα με το όνομα literals
literals = []
movie_score = {}
for row in data.itertuples(index=True, name='Pandas'):
  movie_title = clean_text(getattr(row, 'movie_title'))

  director = getattr(row, 'director_name')
  actor1 = getattr(row, 'actor_1_name')
  actor2 = getattr(row, 'actor_2_name')
  actor3 = getattr(row, 'actor_3_name')
  lang = getattr(row, 'language')
  popularity = getattr(row, 'popularity')
  vote_avg = getattr(row, 'vote_average')
  num_votes = getattr(row, 'num_voted_users')
  duration = getattr(row, 'duration')
  budget = getattr(row, 'budget')
  year = getattr(row, 'title_year')
  country = getattr(row, 'country')
  status = getattr(row, 'status')
  gross = getattr(row, 'gross')
  homepage = getattr(row, 'homepage')


  movie_score[movie_title] = 0
  
  if movie_title != "UNK":

    for genre in getattr(row, 'genres').split("|"):
      literals.append("genre('"+ movie_title +"','"+ clean_text(genre) +"')")

    if director != "UNK":
      literals.append("hasDirector('"+ movie_title +"','"+ clean_text(director) +"')")

    if actor1 != "UNK":
      literals.append("firstActor('"+ movie_title +"','"+ clean_text(actor1) +"')")

    if actor1 != "UNK":
      literals.append("secondActor('"+ movie_title +"','"+ clean_text(actor2) +"')")

    if actor3 != "UNK":
      literals.append("thirdActor('"+ movie_title +"','"+ clean_text(actor3) +"')")

    if lang != "UNK":
      literals.append("spokenLanguage('"+ movie_title +"','"+ clean_text(lang) +"')")

    for keyword in getattr(row, 'plot_keywords').split("|"):
      literals.append("hasPlotKeyword('"+ movie_title +"','"+ clean_text(keyword) +"')")


    if popularity != "UNK":
      literals.append("hasPopularity('"+ movie_title +"','"+ clean_text(str(popularity)) +"')")

    if vote_avg != "UNK":
      literals.append("hasVoteAvg('"+ movie_title +"','"+ str(vote_avg) +"')")

    if num_votes != "UNK":
      literals.append("hasVoters('"+ movie_title +"','"+ str(num_votes) +"')")

    #χρηση της διαρκειας των ταινιων για τη δημιουργια κατηγορήματος σχετικά με το είδος της ταινίας ως προς τη διάρκεια
    #μεγάλου μήκους αν διάρκεια <40 και μικρού αν <=40 σύμφωνα με τους κανόνες του διεθνούς κινηματογράφου
    shortf = "Short film"
    longf = "Long film"
    if duration != "UNK":
      if (duration <=40):
        literals.append("durationType('"+ movie_title +"','"+ shortf +"')")
      else:
        literals.append("durationType('"+ movie_title +"','"+ longf +"')")

    if budget != "UNK":
      literals.append("hasBudget('"+ movie_title +"','"+ str(budget) +"')")

    if year != "UNK":
      literals.append("year('"+ movie_title +"','"+ str(year) +"')")

    if country != "UNK":
      literals.append("country('"+ movie_title +"','"+ clean_text(country) +"')")

    if status != "UNK":
      literals.append("status('"+ movie_title +"','"+ clean_text(status) +"')")

    if gross != "UNK":
      literals.append("gross('"+ movie_title +"','"+ str(gross) +"')")

    if status != "UNK":
      literals.append("homepage('"+ movie_title +"','"+ clean_text(homepage) +"')")


#Η Prolog θέλει τα κατηγορήματά της με την σειρά 
literals.sort()
for literal in literals:
  prolog.assertz(literal)

#1β

#Επίσης μπορούμε να κάνουμε consult ένα έτοιμο αρχείο στον κόσμο όπως παρακάτω
prolog.consult(path + "1_b.pl")
````

````python
from tqdm.notebook import tqdm
from sklearn.metrics import precision_score, recall_score, f1_score


#βάρη βάσει των αξιολογήσεων
rating_weights = { 0: -15, 0.5: -12.5,  1: -10, 1.5: -7.5, 2:-5, 2.5: -2.5, 3: 0, 3.5: 2.5, 4: 5, 4.5: 7.5, 5:10}
#βάρη βάσει των επιπέδων ομοιότητας
score_weights = {i:i + 1 for i in range(65)} # ανάλογα με τα επίπεδα ομοιότητας που έχουν οριστεί στην simple_recommender
similarity_weights = {"similarity_7":64,"similarity_6":32,"similarity_5":16, "similarity_4":8, "similarity_3":4, "similarity_2":2, "similarity_1":1}

def train_recommender(ratings, rating_weights, score_weights, start=0, end = -1, movie_score = {}):
    """
    Στην συνάρτηση αυτή μπορούμε να ορίζουμε ποιο υποσύνολο των αξιολογήσεων θα χρησιμοποιήσουμε για το train,
    ενώ παράλληλα μπορούμε να αρχικοποιήσουμε και το movie_score το οποίο αποθηκεύονται τα score κάθε ταινίας.
    Έτσι αν έχουμε κάνει train τον recommender μας για τις πρώτες 10 ταινίες, μπορούμε να συνεχίσουμε για τις 
    υπόλοιπες χωρίς κάθε φορά να ξανά-διαμορφώνουμε το score για αυτές. Αυτό θα σας βοηθήσει στην εκτέλεση και χρονικά όταν ελέγχετε την 
    απόδοση σε υποσύνολα των ratings π.χ. πρώτες 10 ταινίες έπειτα τις επόμενες 20 κ.ο.κ.
    """
    if end == -1:
        end = len(ratings)

    ratings = ratings.iloc[range(start, end)]
    for row in tqdm(ratings.itertuples(index=True, name='Pandas')):
        movie = clean_text(getattr(row, 'movie_title'))
        rating = getattr(row, 'rating')

        similar_movies = simple_recommender(movie)

        ans7 = {}
        ans6 = {}
        ans5 = {}
        ans4 = {}
        ans3 = {}
        ans2 = {}
        ans1 = {}

        ans7 = similar_movies[0]
        ans6 = similar_movies[1]
        ans5 = similar_movies[2]
        ans4 = similar_movies[3]
        ans3 = similar_movies[4]
        ans2 = similar_movies[5]
        ans1 = similar_movies[6]

        for similar_movie in ans7:
            if similar_movie not in movie_score:
                movie_score[similar_movie] = rating_weights[rating] * score_weights[similarity_weights['similarity_7']]
            else:
                movie_score[similar_movie] += rating_weights[rating] * score_weights[similarity_weights['similarity_7']] #το weight θα το ορίσετε ανα επίπεδο ομοιότητας οι πολύ όμοιες ταινίες θα έχουν μεγαλύτερο βάρος
          
        for similar_movie in ans6:
            if similar_movie not in movie_score:
                movie_score[similar_movie] = rating_weights[rating] * score_weights[similarity_weights['similarity_6']]
            else:
                movie_score[similar_movie] += rating_weights[rating] * score_weights[similarity_weights['similarity_6']] #το weight θα το ορίσετε ανα επίπεδο ομοιότητας οι πολύ όμοιες ταινίες θα έχουν μεγαλύτερο βάρος
          
        for similar_movie in ans5:
            if similar_movie not in movie_score:
                movie_score[similar_movie] = rating_weights[rating] * score_weights[similarity_weights['similarity_5']]
            else:
                movie_score[similar_movie] += rating_weights[rating] * score_weights[similarity_weights['similarity_5']] #το weight θα το ορίσετε ανα επίπεδο ομοιότητας οι πολύ όμοιες ταινίες θα έχουν μεγαλύτερο βάρος
          
        for similar_movie in ans4:
            if similar_movie not in movie_score:
                movie_score[similar_movie] = rating_weights[rating] * score_weights[similarity_weights['similarity_4']]
            else:
                movie_score[similar_movie] += rating_weights[rating] * score_weights[similarity_weights['similarity_4']] #το weight θα το ορίσετε ανα επίπεδο ομοιότητας οι πολύ όμοιες ταινίες θα έχουν μεγαλύτερο βάρος
          
        for similar_movie in ans3:
            if similar_movie not in movie_score:
                movie_score[similar_movie] = rating_weights[rating] * score_weights[similarity_weights['similarity_3']]
            else:
                movie_score[similar_movie] += rating_weights[rating] * score_weights[similarity_weights['similarity_3']] #το weight θα το ορίσετε ανα επίπεδο ομοιότητας οι πολύ όμοιες ταινίες θα έχουν μεγαλύτερο βάρος
          
        for similar_movie in ans2:
            if similar_movie not in movie_score:
                movie_score[similar_movie] = rating_weights[rating] * score_weights[similarity_weights['similarity_2']]
            else:
                movie_score[similar_movie] += rating_weights[rating] * score_weights[similarity_weights['similarity_2']] #το weight θα το ορίσετε ανα επίπεδο ομοιότητας οι πολύ όμοιες ταινίες θα έχουν μεγαλύτερο βάρος
          
        for similar_movie in ans1:
            if similar_movie not in movie_score:
                movie_score[similar_movie] = rating_weights[rating] * score_weights[similarity_weights['similarity_1']]
            else:
                movie_score[similar_movie] += rating_weights[rating] * score_weights[similarity_weights['similarity_1']] #το weight θα το ορίσετε ανα επίπεδο ομοιότητας οι πολύ όμοιες ταινίες θα έχουν μεγαλύτερο βάρος

    print(movie_score)    
    return movie_score


#συνάρτηση για predict
#έχουμε ορίσει ότι μια ταινία θα έπρεπε να είναι προτεινόμενη αν είχε σκορ > -30
#χρήση cnt1 και cnt2 για έλεγχο του πλήθους των ταινιών που έχουν movie_score κι αυτών που δεν έχουν αντίστοιχα για την εξαγωγή συμπερασμάτων σχετικά με τις μετρικές

def predict_example(ratings, movie_score, cnt1=0, cnt2=0):
    real, pred = [], []

    for i, row in enumerate(ratings.itertuples(index=True, name='Pandas')):
        movie = clean_text(getattr(row, 'movie_title'))
        rating = getattr(row, 'rating')

        if movie in movie_score: #αν έχουμε σχηματίσει βαθμολογία για την ταινία αυτή
            pred.append(int(movie_score[movie] > -30.0)) #heuristic για το αν μια ταινία είναι προτεινόμενη
            real.append(int(rating > 3))# έτσι ορίζουμε ότι μια ταινία θα έπρεπε να είναι προτεινόμενη
            #η συνθήκη αυτή δεν μπορεί να αλλάξει
            cnt1=cnt1+1
        else: #δεν μπορούμε να προτείνουμε κάτι για το όποιο δεν έχουμε σχηματίσει εικόνα
            pred.append(0)
            real.append(int(rating > 3))
            cnt2=cnt2+1

        #τα παρακάτω print εξυπηρετούν στην παρατήρηση κι άρα στον προσωπικό έλεγχο των αποτελεσμάτων των μετρικών και στη βελτιστοποίηση αυτών όσο είναι δυνατόν
        #εμφάνιση του τίτλου της ταινίας που εξετάζεται με την αξιολόγηση που δίνεται στο test set
        print(movie)
        print("rating: ", rating)
        #εμφάνιση του αν έχει ή οχι movie_score
        if movie in movie_score:
          print("score: ", movie_score[movie])
        else:
          print("no")
        #εμφάνιση του αντίστοιχου predict και real της τρέχουσας ταινίας για έλεγχο δικό μας περί των σωστών και λανθασμένων προβλέψεων
        print("pred: ", pred[-1])
        print("real: ", real[-1])
        print()
    #εμφάνιση των μετρητών των ταινιών που έχουν ή οχι movie score
    print("cnt1: ", cnt1)
    print("cnt2: ", cnt2)
    
    print(pred)
    print(real)
    return real, pred


def get_metrics(real, pred):
    metrics = {}
    metrics["precision"] = precision_score(real, pred)
    metrics["recall"] = recall_score(real, pred)
    metrics["f1"] = f1_score(real, pred)
    return metrics
````
